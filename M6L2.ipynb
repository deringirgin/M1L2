```python
import tf_keras as keras  # tf-keras iГ§e aktarma вҖ“ .h5 uzantДұlДұ modeller ile uyumlu Keras sГјrГјmГј
from tf_keras.models import load_model  # tf_keras iГ§erisinde load_model fonksiyonunu iГ§e aktarma, bu modeli aГ§mamДұzДұ saДҹlar
from PIL import Image, ImageOps  # Pillow kГјtГјphanesinden Image ve ImageOps sДұnДұflarДұnДұ iГ§e aktarДұyoruz
import numpy as np
```

```python
def detect_bird(model_path, labels_path, image_path):
  np.set_printoptions(suppress=True)
  model = load_model(model_path, compile=False)
  class_names = open(labels_path, "r").readlines()
  data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
  image = Image.open(image_path).convert("RGB")
  size = (224, 224)
  image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)
  image_array = np.asarray(image)
  normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1
  data[0] = normalized_image_array
  prediction = model.predict(data)
  index = np.argmax(prediction)
  class_name = class_names[index]
  confidence_score = prediction[0][index]

  return class_name[2:], confidence_score


className, confidenceScore = detect_bird('/content/keras_model.h5', '/content/labels.txt', '/content/test.png')

if (className == "cat\n"):
  print("Kedi beslemek Г§ok iyidir yaЕҹ ya da kuru mama verebilirsiniz. Bazen Г¶dГјl olarak Г¶dГјl mamasДұ verin")
elif (className == "tiger\n"):
  print("KaplanlarДұ evde beslemek iyi bir fikir deДҹildir ama yavruyken alДұsanДұz size baДҹlanДұrlar beslemek iГ§in et verebilisiniz")
elif (className == "dog\n"):
  print("KГ¶pek iГ§in et ya da mama verilebilir ")
elif (className == "dragon\n"):
  print("Ejderhalar yok ama filmlerde balДұ ya da et yiyorlar")
```
